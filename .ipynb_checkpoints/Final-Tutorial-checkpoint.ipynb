Let's begin by seeing the different levels of pollutants throughout the different neighborhoods. We can accomplish this using Folium, a Python library for creating interactive maps, which will allow us to see the levels of pollutants in each neighborhood and how they vary across the city.

To start and simplify things lets make two functions. The first determines the color of a marker with red being the worst levels, yellow is in the middle and green as the best in the city. The second function does the actual mapping by taking a data frame of a type of pollutant and marking their amounts in the different neighborhoods using the color_producer to help visualize the severity.

```python
# sets the color of each marker depending on where it is in the range of values
def color_producer(ammount, min, max):
    temp = (max-min)/3
    if max-temp < ammount <= max :
        return 'red'
    elif max-(2*temp) < ammount <= max-temp :
        return 'yellow'
    elif min <= ammount <= max-(2*temp):
        return 'green'
    else:
        return 'gray'

def map_maker (df, min, max, colName):
    nyc_map = folium.Map(location=[40.7128, -74.0060], zoom_start=10)
    
    # Add markers for each location
    for index, row in df.iterrows():
        lat = row['Lat']
        lon = row['Long']
        name = row['Geo Place Name']
        folium.CircleMarker(
            location=[lat, lon], 
            radius=7,
            fill = True,
            fill_opacity = 0.5,
            color = color_producer(row[colName],min, max),
            opacity = 0.4,
            popup=name
        ).add_to(nyc_map)
    return nyc_map
```

## Model: Analysis, Hypothesis Testing, & ML
The modeling step is a crucial component of the data science lifecycle. Once the data has been collected, cleaned, and preprocessed, the next step is to create a model that can be used to make predictions or gain a better understanding of the underlying relationships between variables.

While the data that we have so far is good, it does not help to determine how severe the issue is today as it only goes until 2020. Let's create linear regression model to predict the air quality outside the scope of our data.

To begin let's create a function that creates a linear regression model using the ordinary least squares method. OLS finds the line that best fits the data by minimizing the distance between the predicted values and the actual values. The coefficients of the linear equation are then estimated using this line.

```python
def leastSquares (df, X, Y, xlabel, ylabel, title):
    
    # Split the data into training and testing sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)

    # Create a linear regression model
    model = LinearRegression()

    # Train the model on the training data
    model.fit(X_train, Y_train)

    # Compute the predicted values for the training and testing sets
    Y_train_pred = model.predict(X_train)
    Y_test_pred = model.predict(X_test)
    
    print('Ordinary Least Squares (OLS)')
    print('Coefficients: ', model.coef_[0])
    print('Intercept: ', model.intercept_)
    print('Mean squared error: %.2f' % mean_squared_error(Y_test, Y_test_pred))
    print('Coefficient of determination: %.2f' % r2_score(Y_test, Y_test_pred))

    plt.scatter(X_train, Y_train, color='blue', label='Training Data')
    plt.scatter(X_test, Y_test, color='red', label='Testing Data')
    plt.plot(X_train, Y_train_pred, color='black', linewidth=2, label='Regression Line')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.show()
```

Also we can define a function that uses Random Sample Consensus (RANSAC). RANSAC is an iterative algorithm used to estimate the parameters of a mathematical model from a set of data points. It is useful in situations where the data set contains a large number of outliers, which can significantly affect the estimation of the model parameters. By iteratively fitting the model to subsets of the data that contain fewer outliers, RANSAC can provide more robust estimates of the model parameters than traditional methods like least squares regression.

We can see the difference between the two once we start getting the line of regression for the different data sets.
